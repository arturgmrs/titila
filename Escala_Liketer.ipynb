```
# importação dos pacotes numpy & matplotlib
import numpy as np
import matplotlib.pyplot as plt
#
category_name = ['Discordo Totalmente', 'Discordo', 'Indiferente', 
                  'De acordo', 'Totalmente de Concordo']


results = {
    'QP01_______________________________________________________________________________________':[0, 8.3, 16.7, 25, 50],
    'QP02_______________________________________________________________________________________': [0, 0, 0, 25, 75],
    'QP03_______________________________________________________________________________________': [0, 0, 0, 25, 75],
    'QP04_______________________________________________________________________________________': [8.3, 0, 8.3, 33.3, 50],
    'QP05_______________________________________________________________________________________': [0, 0, 0, 25, 75],
    'QP06 - Enquanto jogava você sentiu perder a noção do tempo?______________________________': [0, 8.3, 16.7, 25, 50],
    'QP08 - Você se sentiu desafiado durante o jogo?____________________________________________': [0, 0, 0, 25, 75],
    'QP08 - Você notou a aplicabilidade de alguma competência educacional no jogo?____________': [0, 0, 0, 25, 75],
    'QP09 - Você se divertiu durante o jogo?____________________________________________________': [8.3, 0, 8.3, 33.3, 50],
    'QP10 - Você se divertiu durante o jogo?____________________________________________________': [0, 0, 0, 25, 75],
    'QP11 - Você teve controle durante o jogo?___________________________________________________': [8.3, 8.3, 33.3, 25, 25]
}
def survey(results, category_name):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_name, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts, height=0.5,
                label=colname, color=color)
        xcenters = starts + widths / 2
        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax.legend(ncol=len(category_name), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax

survey(results, category_name)
plt.show()
```
