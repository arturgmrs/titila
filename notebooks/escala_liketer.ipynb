```
# importação dos pacotes numpy & matplotlib
import numpy as np
import matplotlib.pyplot as plt
# Definição das categorias
category_name = ['Discordo Totalmente', 'Discordo', 'Indiferente', 'Concordo', 'Concordo Totalmente']

results = {
    # Pode adicionar a questão ou alterar quantidade de números, em verde são os valores das porcentagens por categoria
    'Questão 01 | Que dia é hoje?': [0, 10, 20, 30, 40],
    'Questão 02 | Você se divertiu durante o jogo?': [0, 0, 0, 25, 75],
    'Questão 03 | Você conseguiu compartilhar os resultados?': [0, 0, 25, 25,50],
    'Questão 04': [8, 1, 8, 33, 50],
    'Questão 05': [0, 0, 0, 25, 75],
    'Questão 06': [0, 8, 1, 21, 70],
    'Questão 07': [0, 0, 0, 25, 75],
    'Questão 08': [0, 0, 0, 25, 75],
    'Questão 09': [8, 1, 8, 33, 50],
    'Questão 10': [0, 0, 0, 25, 75],

}
def survey(results, category_name):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_name, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts, height=0.5,
                label=colname, color=color)
        xcenters = starts + widths / 2
        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax.legend(ncol=len(category_name), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax
    

survey(results, category_name)

plt.show()

#Source: 
```
